```python
cf['Review Rating'] = cf.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
cf.isnull().sum()
```

```python
cf.columns = cf.columns.str.lower()
cf.columns = cf.columns.str.replace(' ','_')
cf = cf.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
cf.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
cf['age_group'] = pd.qcut(cf['age'], q=4, labels = labels)
```

```python
cf[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

cf['purchase_frequency_days'] = cf['frequency_of_purchases'].map(frequency_mapping)
```

```python
cf[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
cf[['discount_applied','promo_code_used']].head(10)
```

```python
# Dropping promo code used column

cf = cf.drop('promo_code_used', axis=1)
```
